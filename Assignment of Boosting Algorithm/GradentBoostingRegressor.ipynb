```python
import pandas as pd
```

```python
dataset = pd.read_csv('50_Startups.csv')
```

```python
dataset = pd.get_dummies(dataset, drop_first = 'true')
```

```python
dataset.columns
```

```python
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'Profit','State_Florida', 'State_New York']]
```

```python
dependent = dataset[['Profit']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.3)
```

```python
param_combination = [
    {},
    {'n_estimators': 100, 'loss': 'squared_error', 'criterion': 'friedman_mse', 'learning_rate': 0.1},
    {'n_estimators': 50, 'loss': 'absolute_error', 'criterion': 'squared_error', 'learning_rate': 0.2},
    {'n_estimators': 10, 'loss': 'huber', 'criterion': 'friedman_mse', 'learning_rate': 0.3},
    {'n_estimators': 1, 'loss': 'quantile', 'criterion': 'squared_error', 'learning_rate': 0.4}
]
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score

for param in param_combination:
    gbr = GradientBoostingRegressor(**param)
    gbr.fit(x_train, y_train)

    y_predict = gbr.predict(x_test)

    result = r2_score(y_test, y_predict)
    print(param, ': r2_score value : ', result)
```
