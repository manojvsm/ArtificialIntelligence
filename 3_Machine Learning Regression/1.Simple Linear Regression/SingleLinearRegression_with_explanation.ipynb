```python
import pandas as pd #pandas is an library it contains function related to tables and databases
```

```python
dataset = pd.read_csv("Salary_Data.csv") #dataset is a variable, using read_csv function in pandas library to read the csv file
```

```python
dataset#retrun the dataset variable values
```

```python
independent = dataset[["YearsExperience"]] #assign the YearsExperience values to the independent variable
dependent = dataset[["Salary"]] #assign the Salary values to the dependent variable
#dependent = dataset["Salary"] #if miss one of the square bracket the below x_train, x_test, y_train, y_test result is different
```

```python
from sklearn.model_selection import train_test_split #import the train_test_split from the sklearn.model_selection library
```

```python
x_train, x_test, y_train, y_test=train_test_split(independent, dependent, test_size = 0.3, random_state = 0) 
#using the train_test_split function to split the data for test and train data, test_size is for how many percenctage of records for test
```

```python
#x_train, x_test, y_train, y_test
```

```python
from sklearn.linear_model import LinearRegression #import the LinearRegression function library from the sklearn.linear_model library
```

```python
regressor = LinearRegression() #LinearRegression() function store to the regressor variable
regressor.fit(x_train, y_train) #using fit function to train the model using the training set of input and output
```

```python
weight = regressor.coef_ #straight line forumla y=mx+c/y=wx+b w-weight b-bias, using coef_ to find the weight of the trained data
```

```python
weight
```

```python
bias = regressor.intercept_ #find the bias of the trained data using intercept_
```

```python
bias
```

```python
#the trained data model to find the output of the test data so we can validate with the original data
y_predict = regressor.predict(x_test)
```

```python
y_predict
```

```python
#import the r2_score for the validating the trained data model, one of the validating parameter is (R-Square) to validate output
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_predict) #compare the test output data to model predicted data
```

```python
r_score #if the output is nearly to 1 then its good model
```

```python
import pickle #import the pickle library for store the model, if we use the pickle library file extension should .sav
```

```python
filename = "FinalModel.sav"
pickle.dump(regressor,open(filename, 'wb')) #open the .sav file and wb comment to write in the sav file
```

```python
loaded_model = pickle.load(open(filename, 'rb')) #read the saved model file load the loaded_model variable
result = loaded_model.predict([[26]]) #using predict function to predict the output for the year of experience salary store in result variable
```
