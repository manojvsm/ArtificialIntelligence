```python
import pandas as pd
```

```python
datasets = pd.read_csv("insurance_pre.csv")
```

```python
datasets = pd.get_dummies(datasets, dtype = int,drop_first = True)
```

```python
datasets.columns
```

```python
independent = datasets[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = datasets['charges']
```

```python
#No need to split the data sets as x_train, y_train, x_test, y_test
#Because we use here as GridSearchCV (CV - Cross Validation)
#Cross validation to all spliting for the datasets
```

```python
from sklearn.model_selection import GridSearchCV 

from sklearn.svm import SVR

estimator = SVR()

param_grid = {
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'gamma': ['scale'],#, 'auto'],
    'degree': [1,2,3]
}

model = GridSearchCV(estimator, param_grid, refit = True)

model
```

```python
model.fit(independent, dependent)
```

```python
model.best_params_
```

```python
#min_samples_split #This will show in dictionary format - just text
result = model.cv_results_
```

```python
pd.DataFrame.from_dict(result) #DataFrame to shown records in table format from the dictionary format
```

```python
age = int(input("Age: "))
bmi = float(input("BMI: "))
children = int(input("Children: "))
sex = bool(input("Sex Male 1 or 0: "))
smoker = bool(input("Smoker 1 or 0: "))

prediction = model.predict([[age, bmi, children, sex, smoker]])

print ("Predict Insurance Amount: {}".format(prediction))
```
