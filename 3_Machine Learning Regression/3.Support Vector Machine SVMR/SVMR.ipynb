```python
import pandas as pd
```

```python
dataset = pd.read_csv('50_Startups.csv')
```

```python
dataset = pd.get_dummies(dataset, dtype = int, drop_first = True)
```

```python
dataset.columns
```

```python
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
dependent = dataset[['Profit']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(independent, dependent,test_size = 0.3, random_state = 0)
```

```python
x_train
```

```python
#Standardisation - convert the raw data to standardisation
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train) #x_train = sc.transform(x_train) without fit throw an error, first have to fit the data then transform
x_test = sc.transform(x_test)
```

```python
x_train, x_test
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = "rbf", C=0.1)#'sigmoid', 'rbf', 'precomputed', 'linear', 'poly' and we can user 'C' C=0.1 or 10 or 100
regressor.fit(x_train, y_train)
```

```python
print (regressor.intercept_,'\n')
print (regressor.support_,'\n')
print (regressor.n_support_,'\n')
```

```python
y_predict = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_predict)
```
