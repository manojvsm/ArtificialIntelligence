```python
from sklearn import datasets
```

```python
dataset = datasets.load_diabetes()
```

```python
independent = dataset.data
```

```python
dependent = dataset.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.3, random_state = 17)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
model = GradientBoostingRegressor()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
param_grid = {
    'n_estimator': [10, 200, 300],
    'learning_rate': [0.1, 0.2, 0.3],
    'max_depth': [3, 5, 6],
    'min_samples_split': [2, 3, 4],
    'min_samples_leaf': [1, 2, 3]
}
```

```python
param_grid
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gbr_cv = GridSearchCV(model, param_grid, cv = 3, n_jobs = -1)
```

```python
gbr_cv.fit(x_train, y_train)
```
