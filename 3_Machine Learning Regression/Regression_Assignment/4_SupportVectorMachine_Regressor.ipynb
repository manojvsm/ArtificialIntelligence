```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
#dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first = 'True', dtype = int)
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.3, random_state = 0)
```

```python
#x_train, x_test, y_train, y_test
```

```python
# Define different parameter combinations
param_combinations = [
    {},#no paramters
    
    {'kernel': 'linear', 'gamma': 'scale', 'epsilon': 0.1},
    {'kernel': 'poly', 'gamma': 'scale', 'epsilon': 0.1},
    {'kernel': 'rbf', 'gamma': 'scale', 'epsilon': 0.1},
    {'kernel': 'sigmoid', 'gamma': 'scale', 'epsilon': 0.1},
    
    {'kernel': 'linear', 'gamma': 'auto', 'epsilon': 0.1},
    {'kernel': 'poly', 'gamma': 'auto', 'epsilon': 0.1},
    {'kernel': 'rbf', 'gamma': 'auto', 'epsilon': 0.1},
    {'kernel': 'sigmoid', 'gamma': 'auto', 'epsilon': 0.1}
]
```

```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score

for param in param_combinations:
    regressor = SVR(**param)
    regressor.fit(x_train, y_train)

    y_predict = regressor.predict(x_test)

    result = r2_score(y_test, y_predict)

    print ('Parameters: ', str(param))
                      
    print ('R2_Score: ',result)
```
