```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
#dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first = 'True', dtype = int)
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.3, random_state = 0)
```

```python
#x_train, x_test, y_train, y_test
```

```python
# Define different parameter combinations
param_combinations = [
    {},#no paramters

    {'criterion':'squared_error', 'splitter':'best','max_features':'sqrt','max_depth' : None},
    {'criterion':'friedman_mse', 'splitter':'best','max_features':'sqrt','max_depth' : None},
    {'criterion':'absolute_error', 'splitter':'best','max_features':'sqrt','max_depth' : None},
    {'criterion':'poisson', 'splitter':'best','max_features':'sqrt','max_depth' : None},
    
    {'criterion':'squared_error', 'splitter':'random','max_features':'sqrt','max_depth' : None},
    {'criterion':'friedman_mse', 'splitter':'random','max_features':'sqrt','max_depth' : None},
    {'criterion':'absolute_error', 'splitter':'random','max_features':'sqrt','max_depth' : None},
    {'criterion':'poisson', 'splitter':'random','max_features':'sqrt','max_depth' : None},

    {'criterion':'squared_error', 'splitter':'best','max_features':'log2','max_depth' : None},
    {'criterion':'friedman_mse', 'splitter':'best','max_features':'log2','max_depth' : None},
    {'criterion':'absolute_error', 'splitter':'best','max_features':'log2','max_depth' : None},
    {'criterion':'poisson', 'splitter':'best','max_features':'log2','max_depth' : None},

    {'criterion':'squared_error', 'splitter':'random','max_features':'log2','max_depth' : None},
    {'criterion':'friedman_mse', 'splitter':'random','max_features':'log2','max_depth' : None},
    {'criterion':'absolute_error', 'splitter':'random','max_features':'log2','max_depth' : None},
    {'criterion':'poisson', 'splitter':'random','max_features':'log2','max_depth' : None}
]
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

#for tree image package
#import matplotlib.pyplot as plt
#from sklearn import tree

for param in param_combinations:
    regressor = DecisionTreeRegressor(**param)
    regressor.fit(x_train, y_train)

    y_predict = regressor.predict(x_test)

    result = r2_score(y_test, y_predict)

    print ('Parameters: ', str(param))
                      
    print ('R2_Score: ',result)

    #shown image in tree structure
    #tree.plot_tree(regressor)
    #plt.show()
```
