```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
#dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first = 'True', dtype = int)
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.3, random_state = 0)
```

```python
#x_train, x_test, y_train, y_test
```

```python
# Define different parameter combinations
param_combinations = [
    {},#no paramters
    
    {'n_estimators':100, 'criterion':'squared_error','random_state' : None,'max_features':'sqrt'},
    {'n_estimators':50, 'criterion':'squared_error','random_state' : None,'max_features':'log2'},
    {'n_estimators':10, 'criterion':'squared_error','random_state' : None,'max_features':None},

    {'n_estimators':100, 'criterion':'absolute_error','random_state' : None,'max_features':'sqrt'},
    {'n_estimators':50, 'criterion':'absolute_error','random_state' : None,'max_features':'log2'},
    {'n_estimators':10, 'criterion':'absolute_error','random_state' : None,'max_features':None},

    {'n_estimators':100, 'criterion':'friedman_mse','random_state' : None,'max_features':'sqrt'},
    {'n_estimators':50, 'criterion':'friedman_mse','random_state' : None,'max_features':'log2'},
    {'n_estimators':10, 'criterion':'friedman_mse','random_state' : None,'max_features':None},

    {'n_estimators':100, 'criterion':'poisson','random_state' : None,'max_features':'sqrt'},
    {'n_estimators':50, 'criterion':'poisson','random_state' : None,'max_features':'log2'},
    {'n_estimators':10, 'criterion':'poisson','random_state' : None,'max_features':None} 
]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

#for tree image package
#import matplotlib.pyplot as plt
#from sklearn import tree

for param in param_combinations:
    regressor = RandomForestRegressor(**param)
    regressor.fit(x_train, y_train)

    y_predict = regressor.predict(x_test)

    result = r2_score(y_test, y_predict)

    print ('Parameters: ', str(param))
                      
    print ('R2_Score: ',result)

    #shown image in tree structure
    #tree.plot_tree(regressor)
    #plt.show()
```
